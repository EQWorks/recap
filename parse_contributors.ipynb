```python
import pandas as pd
```

```python
# using get-contributors.ps1 PowerShell script
df = pd.read_csv('./2021/contributors.csv')
```

```python
filtered = df[df.Commits > 0].reset_index(drop=True)
```

```python
len(filtered.Author.unique())
```

```python
by_repo = filtered.groupby('Repository').aggregate({'Author': ['unique'], 'Added': ['sum', 'mean'], 'Deleted': ['sum', 'mean'], 'Commits': ['sum', 'mean']})
by_repo
```
