```python
import itertools

# get word cloud based on T1 and T2
tags = pd.DataFrame({
    'name': itertools.chain(*[
        df[k].dropna().tolist()
        for df in (prs, issues)
        for k in ('t1', 't2')
    ])
})
tags = tags[tags.name != 'others']
tags
```

```python
by_name = flatten_gb(tags.groupby('name').aggregate({'name': ['count']}))
by_name
```

```python
by_name = by_name.sort_values(by='name_count', ascending=False).set_index('name')
by_name
```

```python
by_name.to_csv('./2021/tags.csv') # output as source to generate a tag/word cloud
```

```python
# sample tag/word cloud
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS

wc = WordCloud(
    width = 900,
    height = 900,
    mode='RGBA',
    background_color = None,
    stopwords = STOPWORDS,
)
wc.generate_from_text(' '.join(tags.name.values.tolist()))
fig = plt.figure(
    figsize = (40, 30),
    facecolor = 'k',
    edgecolor = 'k')
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.savefig('./wc.png')
plt.show()
```

```python
# sample masked cloud
from PIL import Image
import numpy as np

mask = np.array(Image.open('./bender.jpg'))
wc = WordCloud(
    width = 900,
    height = 900,
    mode = 'RGBA',
    background_color = None,
    stopwords = STOPWORDS,
    mask=mask,
)
wc.generate_from_text(' '.join(tags.name.values.tolist()))
fig = plt.figure(
    figsize = (10, 10),
    # facecolor = 'k',
    # edgecolor = 'k',
)
plt.imshow(wc, interpolation = 'bilinear')
plt.axis('off')
plt.tight_layout(pad=0)
plt.savefig('./wc_bender.png')
plt.show()
```
