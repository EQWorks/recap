```python
import os
import pandas as pd

# CSVs from AWS Console monthly billing download
for parent, dirs, files in os.walk('./2021/aws/'):
    data = pd.concat([pd.read_csv(os.path.join(parent, f)) for f in files], ignore_index=True)
    break
```

```python
data.columns
```

```python
df = data[['BillingPeriodStartDate', 'BillingPeriodEndDate', 'ProductCode', 'ProductName', 'UsageType', 'ItemDescription', 'UsageQuantity', 'CostBeforeTax', 'Credits', 'TaxAmount', 'TotalCost']]
df
```

```python
df = df.dropna()
df
```

```python
from utils import flatten_gb
```

```python
by_date_product = flatten_gb(df.groupby(['ProductName', 'BillingPeriodStartDate', 'UsageType', 'ItemDescription']).aggregate({
    'UsageQuantity': ['sum', 'mean'],
    'TotalCost': ['sum', 'mean'],
}))
by_date_product
```

```python
by_product = flatten_gb(df.groupby(['ProductName', 'UsageType', 'ItemDescription']).aggregate({
    'UsageQuantity': ['sum', 'mean'],
    'TotalCost': ['sum', 'mean'],
}))
by_product
```

```python
s3 = by_product[by_product['ProductName'] == 'Amazon Simple Storage Service'].reset_index(drop=True)
s3
```

```python
def extract_s3_type(UsageType):
    _type = UsageType

    for reg in ('CAN1', 'USW1', 'USE2', 'USW2'):
        _t = _type.split('-')
        if _t[0] == reg:
            _type = '-'.join(_t[1:])
            break

    _type = _type.split('-')[0]
    return _type

s3['type'] = s3.UsageType.apply(extract_s3_type)
s3
```

```python
s3_by_type = s3.groupby('type').aggregate({
    'ItemDescription': 'unique',
    'UsageQuantity_sum': 'sum',
    'TotalCost_sum': 'sum',
})
s3_by_type
```

```python
ec2 = by_product[by_product['ProductName'] == 'Amazon Elastic Compute Cloud'].reset_index(drop=True)
ec2
```

```python
def extract_ec2_type(UsageType):
    _type = UsageType

    for reg in ('CAN1', 'USW1', 'USE1', 'USE2', 'USW2'):
        _t = _type.split('-')
        if _t[0] == reg:
            _type = '-'.join(_t[1:])
            break

    _type = ':'.join(_type.split(':')[0:-1])
    return _type

ec2['type'] = ec2.UsageType.apply(extract_ec2_type)
ec2
```

```python
ec2_by_type = ec2.groupby('type').aggregate({
    'ItemDescription': 'unique',
    'UsageQuantity_sum': 'sum',
    'TotalCost_sum': 'sum',
})
ec2_by_type
```
